```python
import pandas as pd
```

```python
# Import complete accident information dataset
file_1 = "Resources/Accident_Information.csv"
accident_df = pd.read_csv(file_1)

accident_df.head()
```

```python
# select required columns
columns = ['Accident_Index', '1st_Road_Class','Accident_Severity','Carriageway_Hazards','Year','Date','Day_of_Week','Did_Police_Officer_Attend_Scene_of_Accident','Junction_Control','Junction_Detail','Junction_Detail','Latitude','Light_Conditions','Local_Authority_(District)','Local_Authority_(Highway)','Longitude','Number_of_Casualties','Number_of_Vehicles','Road_Surface_Conditions','Road_Type','Special_Conditions_at_Site','Speed_limit','Time','Urban_or_Rural_Area','Weather_Conditions']
           
acc_df1 = pd.DataFrame(accident_df, columns=columns)
acc_df1           
```

```python
#review data types
acc_df1.info()
```

```python
acc_df1.count()
```

```python
#remove data with misssing values
acc_df2=acc_df1.dropna(how='any')
acc_df2
```

```python
#check to see if dropna worked
acc_df2.count()
```

```python
#repeat the process with the second dataset
file_2 = "Resources/Vehicle_Information.csv"
vehicle_df = pd.read_csv(file_2)

vehicle_df.head()
```

```python
#repeat the process with the second dataset
columns1 = ['Accident_Index', 'Age_Band_of_Driver','Age_of_Vehicle','Driver_Home_Area_Type','Engine_Capacity_.CC.','Hit_Object_in_Carriageway','Hit_Object_off_Carriageway','Journey_Purpose_of_Driver','Junction_Location','make','model','Propulsion_Code','Sex_of_Driver','Towing_and_Articulation','Vehicle_Manoeuvre','Vehicle_Type','Was_Vehicle_Left_Hand_Drive','X1st_Point_of_Impact']
           
veh_df1 = pd.DataFrame(vehicle_df, columns=columns1)
veh_df1    
```

```python
#repeat the process with the second dataset
veh_df1.info()
```

```python
#repeat the process with the second dataset
veh_df1.count()
```

```python
#repeat the process with the second dataset
veh_df2=veh_df1.dropna(how='any')
veh_df2
```

```python
#repeat the process with the second dataset
veh_df2.count()
```

```python
#review data types
veh_df2.info()
```

```python
#merge the datasets using an outer merge
df2_merge = acc_df2.merge(veh_df2, on="Accident_Index", how = 'outer')
df2_merge
```

```python
#review the data types and columns
df2_merge.info()
```

```python
#remove any rows with missing or values
df2_drop=df2_merge.dropna(how='any')
df2_drop
```

```python
#review types
df2_drop.info()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
df2_drop.to_csv("Resources/df2.csv", index=False, header=True)
```
