```python
# Dependencies
import gmaps
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import pearsonr
```

```python
# Read main CSV data file
accidents_df = pd.read_csv("all.csv", low_memory=False)

# Print the columns list to see available columns
print(accidents_df.columns)

# Get an overview of the DataFrame
accidents_df
```

I have imported the cleaned data from the CSV file where it was stored.

```python
# Group by the Road Class into new DataFrame
class_group = accidents_df.groupby("1st_Road_Class")

# Count the different accident severity types by the road_class group
class_severity_count = class_group["Accident_Severity"].value_counts()

# Create a DataFrame to plot
road_class_df = pd.DataFrame({"Accident Severity": class_severity_count})

# Sort the DataFrame by Largest to smallest number of Accidents
road_class_df = road_class_df.sort_values("Accident Severity", ascending=False)

# Rename the Accident Severity to Number of Accidents column as it is a count.
road_class_df = road_class_df.rename(columns={"Accident Severity":"Number of Accidents"})

# Plot the DataFrame as bar plot
road_class_df.plot(kind="bar")

# Plot Title and Axis Labels
plt.title("Number of Accidents by Severity for different Road Classes")
plt.xlabel("Road Class")
plt.ylabel("Number of Accidents")

# Show the resulting DataFrame
plt.show()
```

Plot appears to indicate the overwhelming majority of accidents take place on A Roads and have a severity rating of slight. The A road also have the most number Fatal accidents. This appears to be an indication that travelling on A roads puts you at the greatest chance of being involved in accident.
