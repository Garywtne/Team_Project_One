```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import scipy.stats as sts
from scipy.stats import linregress
import numpy as np
import seaborn as sns
import scipy.stats as st
```

```python
accident_df = pd.read_csv("all.csv", low_memory=False)

accident_df.head()
```

```python
all_average = accident_df.groupby(["Speed_limit"]).mean()
all_average.sort_values(by=["Number_of_Casualties"], ascending=False)
average_casualties = all_average["Number_of_Casualties"]
len(average_casualties)
```

```python
all_average.
```

```python


corr=round(st.pearsonr(all_average[""],average_casualties)[0],2)
print(f"The correlation is {corr}")
```
